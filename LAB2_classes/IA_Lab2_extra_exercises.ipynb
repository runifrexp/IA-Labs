A product has the following characteristics:
*   id: unique numerical identifier of the product
*   name: name of the product
*   price: price of the product

Add a method print to print the attributes of the Product

```python
# code here the class Product considering the attributes mentioned before

class Product:
  def __init__(self, id: int, name: str, price: float):
    self.id = id
    self.name = name
    self.price = price

  def print_product(self):
    print('ID: ' + str(self.id)+
          '\nname: ' + self.name +
          '\nprice: ' + str(self.price) + '€')
```

Now write the class PriceList that manages a shop's price list.

This class contains instances of the Product class (the class just written).

Furthermore, the PriceList class must allow the following functionalities:

*  add Product in the PriceList
*  find the price of a Product from its id
*  compute the total value of the PriceList (given by the sum of the prices of each product contained in the PriceList)

```python
# code here the class PriceList considering the functionalities mentioned before

class Pricelist(Product):
  def __init__(self):
    self.lista = []

  def __add__(self, otherProduct):
    self.lista.append(otherProduct)
    return self

  def find(self, id):
    for product in self.lista:
      if id == product.id:
        return product.price
    return None

  def total_price(self):
    tot = 0
    for product in self.lista:
      tot += product.price
    return tot

  def print_Pricelist(self):
    i = 0
    for product in self.lista:
      print("\nprodotto " + str(i))
      product.print_product()
      i += 1
```

```python
# add a main to check if all is working properly

def main():
  print("\n")
  p1 = Product(1, "foglie", 1.2)
  p2 = Product(2, "rocce", 3.6)
  p3 = Product(3, "rame", 6.3)
  p1.print_product()     #print_product works
  print("\n-------------------------------------------------")

  listaprezzi = Pricelist()
  listaprezzi + p1 + p2 + p3      #__add__ works
  listaprezzi.print_Pricelist()    #print_Pricelist works
  print("\n-------------------------------------------------\n")

  price_2 = listaprezzi.find(2)
  print(price_2)      #find works
  print("\n-------------------------------------------------\n")

  tot = listaprezzi.total_price()       #total_price works
  print("total price: " + str(tot) + "€")
  print("\n")

main()
```

In addition to the Price List class, the Cart class also uses the Product class.

The Cart class has the following functionalities:
*   add a Product with the relative quantity
*   compute the total value of the Cart (given by the sum of the prices of each product contained in the Cart)

```python
# code here the class Cart considering the functionalities mentioned before

class Cart(Pricelist):
  def __init__(self):
    self.lista = list()

  def __add__(self, otherProduct):
    self.lista.append(otherProduct)
    return self

  def total_Cart(self):
    tot = 0
    for (product, quantity) in self.lista:
      tot += product.price * quantity
    return tot
```

```python
# add a main to check if all is working properly

def main():
  p1 = Product(1, "foglie", 1.2)
  p2 = Product(2, "rocce", 3.6)
  p3 = Product(3, "rame", 6.3)
  carrello = Cart()
  carrello + (p1, 3) + (p2, 5) + (p3, 2)       #__add__ works
  tot = carrello.total_Cart()
  print("\ntotal of the cart: " + str(tot) + "€\n")       #total_cart works
main()
```

Starting from the list old_list = [[1, 1, 1], [2, 2, 2], [3, 3, 3]], create a shallow copy and then modify the list in order to store [[1, 1, 1], [2, 2, 'BB'], [3, 3, 3]]

```python
# Use the module copy

import copy

old_list = [[1, 1, 1], [2, 2, 2], [3, 3, 3]]

new_list = old_list.copy()

new_list[1][2] = 'BB'

print(new_list)
```

**Starting** from the list old_list = [[1, 1, 1], [2, 2, 2], [3, 3, 3]], create a deep copy and then modify the list in order to store [[1, 1, 1], [2, 2, 'BB'], [3, 3, 3]]

```python
# Use the module copy

import copy

old_list = [[1, 1, 1], [2, 2, 2], [3, 3, 3]]

new_list = copy.deepcopy(old_list)

new_list[1][2] = 'BB'

print("Modified deep copy:", new_list)
print("\nOriginal list:", old_list)
```

Write a Python class Restaurant with attributes:
*   menu_items
*   booked_table
*   customer_orders

and methods:
*   add_item_to_menu(item, price)
*   book_tables(table)
*   customer_order(table, item)

In the main, perform the following tasks now:

    Now add items to the menu.
    Make table reservations.
    Take customer orders.
    Print the menu.
    Print table reservations.
    Print customer orders.

Note: Use dictionaries and lists to store the data.

```python
class Restaurant:
  def __init__(self):
    self.menu_items = dict()
    self.booked_table = []
    self.customer_orders = dict()

  def add_item_to_menu(self, item, price):
    self.menu_items[item] = price

  def book_tables(self, table):
    self.booked_table.append(table)

  def customer_order(self, table, item):
    if table not in self.customer_orders:
      self.customer_orders[table] = []
    self.customer_orders[table].append(item)

  def print_menu(self):
    print('******MENÙ******')
    for item, price in self.menu_items.items():
      print("{:<12} {}€".format(item, price))
      #print(item + ' ' + str(price) + '€\n')

  def print_booked_tables(self):
    print('******BOOKED TABLES******')
    for table in self.booked_table:
      print('table ' + str(table))

  def print_customer_orders(self):
    print('******CUSTOMER ORDERS******')
    for table, items in self.customer_orders.items():
      print('Table ' + str(table) + ': ' + ', '.join(items))

def main():
  restaurant = Restaurant()

  restaurant.add_item_to_menu('spaghetti', 10)
  restaurant.add_item_to_menu('canederli', 14)
  restaurant.add_item_to_menu('zuppa', 8)
  restaurant.add_item_to_menu('lasagne', 13)
  restaurant.add_item_to_menu('tortellini', 12)
  restaurant.add_item_to_menu('pizza', 7)

  restaurant.book_tables(2)
  restaurant.book_tables(12)
  restaurant.book_tables(15)
  restaurant.book_tables(3)
  restaurant.book_tables(7)

  restaurant.customer_order(2, 'pizza')
  restaurant.customer_order(3, 'spaghetti')
  restaurant.customer_order(12, 'canederli')

  restaurant.print_menu()

  print('\n')
  restaurant.print_booked_tables()

  print('\n')
  restaurant.print_customer_orders()

main()


```
