## EXERCISE 1: Weather's probability
You are given a (fake) <a href="https://drive.google.com/file/d/1LjZLE9ozaHcBwiCl90mHaS1nXKcglfr4/view">padua_weather.csv</a>
of historical records for Padua's weather. The weather, which can be either rainy (= 1 in the dataset), misty (= 2), or sunny (= 3), is reported for each day of the week, for a whole year (52 weeks).

After you formalised the problem (i.e. identify the random variables and necessary mathematical formulae), write a Python program that reads the dataset via Python code and computes the following:
- probability of being sunny during the weekend (one or both days);
- expected weather for each day of the week (*);
- supposed you don't know which day of the week is today: although very unrealistic, how could you guess which day is today based only on the weather?

(\*) An expected value of, for example, 2.5 can be interpreted as "a mix of misty and sunny weather".




```python
import pandas as pd
url = "https://drive.google.com/uc?export=download&id=1LjZLE9ozaHcBwiCl90mHaS1nXKcglfr4"
weather = pd.read_csv(url)
print(weather)
```

probability of being sunny during the weekend (one or both days)

```python
tot = 52 #total weekends

sat = sum(weather['Saturday'] == 3)
sun = sum(weather['Sunday'] == 3)
sat_sun = sum((weather['Saturday'] == 3) & (weather['Sunday'] == 3))

#or
sunny_weekend_count = weather[(weather['Saturday'] == 3) |
                              (weather['Sunday'] == 3)].shape[0]

sunny = sat + sun - sat_sun

print(sat)
print(sun)
print(sat_sun,'\n')
print(sunny/tot)
print(sunny_weekend_count/tot)
```

expected weather for each day of the week.
An expected value of, for example, 2.5 can be interpreted as "a mix of misty and sunny weather".

```python
means = weather.mean()
print(means, '\n')
#means are not good for discrete variables, mode is better, see computational
#statistics lecture 28-03-2024

mon_sunny = sum(weather['Monday'] == 1)
mon_misty = sum(weather['Monday'] == 2)
mon_rainy = sum(weather['Monday'] == 3)
print('sunny mondays =', mon_sunny,
      '\nmisty mondays =', mon_misty,
      '\nrainy mondays =', mon_rainy, '\n')

#or
weather_counts_monday = weather['Monday'].value_counts()
print(weather_counts_monday)
```

supposed you don't know which day of the week is today: although very unrealistic, how could you guess which day is today based only on the weather?

Given the weather, I could guess the day by selecting the nearest occurrencies to that, for example if it is given it's rainy, i'll have to check the day that counts the most rainy days and choose it.

```python
for day in ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']:
    print(f"Occorrenze del meteo per {day}:")
    print(weather[day].value_counts())
    print()  # Linea vuota per separare i risultati
```

so given it's rainy, the day in probably Sunday
